```python
# write full image
def writeFile1(fileName, scalarDataName):
    data = '''# vtk DataFile Version 2.0\nVolume example\nASCII\nDATASET STRUCTURED_POINTS\nDIMENSIONS {} {} {}\nASPECT_RATIO 0.93 0.93 5\nORIGIN 0 0 0\nPOINT_DATA {}\nSCALARS {} float 1\nLOOKUP_TABLE default\n'''.format(xn, yn, zn, xn * yn * zn, scalarDataName)
    tD = [0 for i in range(xn * yn * zn)]
    tN = 0
    tVN = 0
    for z in range(zn):
        for y in range(yn):
            for x in range(xn):
                tD[tN] = "{:.2f}".format(d[x][y][z])
                tN = tN + 1
    data = data + " ".join(tD)
    f = open(fileName, "w")
    f.write(data)
    f.close()
# writeFile1("D:/projects/finalProject/nifti/" + no + "/image.vtk", "image")
```

```python
# write mask
def writeFile(fileName, scalarDataName):
    data = '''# vtk DataFile Version 2.0\nVolume example\nASCII\nDATASET STRUCTURED_POINTS\nDIMENSIONS {} {} {}\nASPECT_RATIO 0.93 0.93 5\nORIGIN 0 0 0\nPOINT_DATA {}\nSCALARS {} float 1\nLOOKUP_TABLE default\n'''.format(xn, yn, zn, xn * yn * zn, scalarDataName)
    tD = [0 for i in range(xn * yn * zn)]
    tN = 0
    tVN = 0
    for z in range(zn):
        for y in range(yn):
            for x in range(xn):
                tD[tN] = "{:.2f}".format(d2[x][y][z])
                tN = tN + 1
    data = data + " ".join(tD)
    f = open(fileName, "w")
    f.write(data)
    f.close()
# writeFile("D:/projects/finalProject/nifti/" + no + "/mask.vtk", "mask")
```

```python
# write full in range
def writeFile2(fileName, scalarDataName):
    data = '''# vtk DataFile Version 2.0\nVolume example\nASCII\nDATASET STRUCTURED_POINTS\nDIMENSIONS {} {} {}\nASPECT_RATIO 0.93 0.93 5\nORIGIN 0 0 0\nPOINT_DATA {}\nSCALARS {} float 1\nLOOKUP_TABLE default\n'''.format(xn, yn, zn, xn * yn * zn, scalarDataName)
    tD = [0 for i in range(xn * yn * zn)]
    tN = 0
    tVN = 0
    for z in range(zn):
        for y in range(yn):
            for x in range(xn):
                if d[x][y][z] >= -879 and d[x][y][z] <= 73:
                    tD[tN] = "{:.2f}".format(d[x][y][z])
                else:
                    tD[tN] = "{:.2f}".format(-880)
                tN = tN + 1
    data = data + " ".join(tD)
    f = open(fileName, "w")
    f.write(data)
    f.close()
# writeFile2("D:/projects/finalProject/nifti/" + no + "/imageFullRange.vtk", "imageFullRange")
```

```python
def unique(list1):
    uniqueList = [ ]
    tMin = float("inf")
    tMax = float("-inf")
    seqMin = list1[0]
    seqMax = list1[0]
    seqList = [ ]
    for x in list1:
        if x not in uniqueList:
            uniqueList.append(x)
        if x < tMin:
            tMin = x
        if x > tMax:
            tMax = x
        if (seqMin - 1) <= x and (seqMax + 1) >= x and seqMin > x:
            seqMin = x
        elif (seqMin - 1) <= x and (seqMax + 1) >= x and seqMax < x:
            seqMax = x
        elif (seqMin - 1) <= x and (seqMax + 1) >= x:
            continue
        else:
            seqList.append((seqMin, seqMax, seqMax - seqMin))
            seqMin = x
            seqMax = x
    return [uniqueList, tMin, tMax, seqList]
def getOrgan(findFor):
    txMin = xn - 1
    txMax = 0
    tyMin = yn - 1
    tyMax = 0
    tzMin = zn - 1
    tzMax = 0
    tData = [ ]
    for z in range(zn):
        for y in range(yn):
            for x in range(xn):
                if d2[x][y][z] == findFor:
                    tData.append(d[x][y][z])
                if d2[x][y][z] == findFor and txMin > x:
                    txMin = x
                if d2[x][y][z] == findFor and txMax < x:
                    txMax = x
                if d2[x][y][z] == findFor and tyMin > y:
                    tyMin = y
                if d2[x][y][z] == findFor and tyMax < y:
                    tyMax = y
                if d2[x][y][z] == findFor and tzMin > z:
                    tzMin = z
                if d2[x][y][z] == findFor and tzMax < z:
                    tzMax = z
    tData.sort()
    return [tData, txMin, txMax, tyMin, tyMax, tzMin, tzMax]
# organ1 = getOrgan(1)
# organData1 = unique(organ1[0])
# organ2 = getOrgan(2)
# organData2 = unique(organ2[0])
# organ3 = getOrgan(3)
# organData3 = unique(organ3[0])

# print("total organ area ({} to {})".format(min(organData1[1], organData2[1], organData3[1]), max(organData1[2], organData2[2], organData3[2])))
# print("(x: {}, {}); (y: {}, {}); (z: {}, {})".format(min(organ1[1], organ2[1], organ3[1]), max(organ1[2], organ2[2], organ3[2]), min(organ1[3], organ2[3], organ3[3]), max(organ1[4], organ2[4], organ3[4]), min(organ1[5], organ2[5], organ3[5]), max(organ1[6], organ2[6], organ3[6])))
# print("organ 1 ({} to {})".format(organData1[1], organData1[2]))
# print("(x: {}, {}); (y: {}, {}); (z: {}, {})".format(organ1[1], organ1[2], organ1[3], organ1[4], organ1[5], organ1[6]))
# print("organ 2 ({} to {})".format(organData2[1], organData2[2]))
# print("(x: {}, {}); (y: {}, {}); (z: {}, {})".format(organ2[1], organ2[2], organ2[3], organ2[4], organ2[5], organ2[6]))
# print("organ 3 ({} to {})".format(organData3[1], organData3[2]))
# print("(x: {}, {}); (y: {}, {}); (z: {}, {})".format(organ3[1], organ3[2], organ3[3], organ3[4], organ3[5], organ3[6]))
```
