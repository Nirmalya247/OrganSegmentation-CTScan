```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
no = "100"
# example_filename = os.path.join(data_path, 'example4d.nii.gz')
img = nib.load('D:/projects/finalProject/nifti/' + no + '/image.nii.gz')
img2 = nib.load('D:/projects/finalProject/nifti/' + no + '/mask.nii.gz')
xn = img.shape[0]
yn = img.shape[1]
zn = img.shape[2]
d = img.get_data()
d2 = img2.get_data()
```

```python
# write full image
def writeFile1(no, scalarDataName, z, d3, minVal = -2000, maxVal = 2000):
    data = '''# vtk DataFile Version 2.0\nVolume example\nASCII\nDATASET STRUCTURED_POINTS\nDIMENSIONS {} {} 1\nASPECT_RATIO 0.93 0.93 1\nORIGIN 0 0 0\nPOINT_DATA {}\nSCALARS {} float 1\nLOOKUP_TABLE default\n'''.format(xn, yn, xn * yn, scalarDataName)
    tD = [0 for i in range(xn * yn)]
    tN = 0
    tVN = 0
    for y in range(yn):
        for x in range(xn):
            if d3[x][y][z] > minVal and d3[x][y][z] < maxVal:
                tD[tN] = "{:.2f}".format(d3[x][y][z])
            elif d3[x][y][z] <= minVal:
                tD[tN] = "{:.2f}".format(minVal)
            else:
                tD[tN] = "{:.2f}".format(maxVal)
            tN = tN + 1
    data = data + " ".join(tD)
    zs = str(z)
    if z < 100:
        zs = "0{}".format(zs)
    f = open("D:/projects/finalProject/nifti/{}/slice/{}{}.vtk".format(no, scalarDataName, zs), "w")
    f.write(data)
    f.close()
```

```python
for slice in range(40, 59):
    writeFile1(no, "raw", slice, d)
    writeFile1(no, "mark", slice, d2)
```

```python
for slice in range(74, 106):
    writeFile1(no, "rawRanged", slice, d, -150, 100)
```
